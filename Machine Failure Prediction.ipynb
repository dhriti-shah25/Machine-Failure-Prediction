```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression  #to use logistic regression model
from sklearn.metrics import accuracy_score
```

```python
#reading the data file
machine_data = pd.read_csv('data.csv')
machine_data
```

```python
#Separating data and labels
# axis = 1 if dropping column , axis = 0 if dropping a row
x = machine_data.drop(columns = 'fail', axis = 1)
y = machine_data['fail']
print(x)
print(y)
```

```python
# Data Standardization
scaler =  StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x)
print(standardized_data)
```

```python
# split data into training data and split data
# test_size 0.2 means 20% data is test data rest train
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model Training ---> Logistics regression model
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#model evaluation
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)#comparing prediction w original label
print("Training data accuracy :",training_data_accuracy)

#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print("Test data accuracy :",test_data_accuracy)
```

```python
# The train_data and test_data accuracy are both high (around 91%). Hence we can move ahead with making a predictive system
```

```python
#Making a predictive system because accuracy score is good
input_data = (190,1,3,3,5,1,20,4,1) #any data set values
# changing input as numpy array
input_data_as_numpy_array = np.asarray(input_data)
# reshaping the array as we are predicting only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
# standardize the input data
std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 0): #prediction is a list hence prediction[0]
    print("The machine will not undergo failure")
else:
    print("The machine will undergo failure")
```
